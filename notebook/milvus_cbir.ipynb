## Introduction

The challenge was to create an end-to-end application scenario using Milvus. Using inspiration from David Shvimer's Towards Data Science post (https://towardsdatascience.com/up-and-running-with-milvus-2466161e8b1f), I decided to use Milvus to detect vector similarity for images, primarily that of cats and dogs. 

Like in the article mentioned, I decided first to use transfer learning for feature extraction and use the features from transfer learning to compute cosine similarities between the images. Then I input the feature vectors into Milvus.

The data is linked here: https://www.kaggle.com/c/dogs-vs-cats/data?select=test1.zip where I only used a subset of  'test1.zip'.

## Transfer Learning

I used ResNet50 and in Tensorflow rather than PyTorch. I chose ResNet50 as I saw that ResNet is quite popular to use for feature extraction, and after pooling, the output is a flat vector of length 2048. Resnet50 is already pre-trained on ImageNet, and I did not fine-tune the model for this task. Instead, I defined a GlobalAveragePooling layer and extracted the feature vectors for my images from there. These features were then compared using cosine similarity.

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import tensorflow as tf
import keras
import os
```

```python
# load in images from directory
images = []
for image in os.listdir('test'):
    im = Image.open('test/' + image)
    images.append(np.array(im))
```

```python
# resize images for input to ResNet
resized = [tf.expand_dims(tf.image.resize(i, (224, 224)), 0) for i in images]
```

```python
# import ResNet50, include_top is False because we do not need the dense layer
base_model = tf.keras.applications.ResNet50(include_top=False, weights='imagenet')
```

```python
# create pooling layer and add on top of base model
global_average_layer = tf.keras.layers.GlobalAveragePooling2D()
model = tf.keras.Sequential([
  base_model,
  global_average_layer])

model.summary()
```

```python
# get feature vectors for each image
feature_vecs = []
for i in range(len(resized)):
    feature_vecs.append(model(resized[i]).numpy())
```

```python
# functions to find top cosine similarities
from sklearn.metrics.pairwise import cosine_similarity
def cos_similarity(idx1, idx2):
    print(cosine_similarity(feature_vecs[idx1], feature_vecs[idx2])[0][0])
    fig, ax = plt.subplots(1, 2, figsize=(10,5))
    ax[0].imshow(images[idx1])
    ax[1].imshow(images[idx2])
    plt.tight_layout()

# idx1: idx of image of interest
# top_k: top k cosine similarities
def top_similarity(top_k, idx1):
    sim_dict={}
    for i in range(len(feature_vecs)):
        # compute cosine similarity between idx1 and each feature vector
        # output sorted by descending
        sim_dict[i]= cosine_similarity(feature_vecs[idx1], feature_vecs[i])[0][0]
    return dict(sorted(sim_dict.items(), key=lambda item: item[1])[::-1][:top_k])

cos_results = top_similarity(10, -4)
```

```python
# plot out similarity results for image of interest
for key, val in cos_results.items():
    plt.figure()
    plt.subplot(1, 2, 1)
    plt.imshow(images[-4])
    plt.subplot(1, 2, 2)
    plt.imshow(images[key])
    print('Cos Similarity: {}'.format(val))
```

As you can see, the cosine similarities are quite mixed. The similarities start to be around the same after the second entry, and even the third most similar image is a dog. Let's see if Milvus can do better.

## Milvus

```python
# connect to server
# docker env
from milvus import Milvus, IndexType, MetricType, Status
_HOST = 'milvus'
_PORT = '19530'
milvus = Milvus(_HOST, _PORT, pool_size=10)
```

```python
# define collection name and set dimension size to 2048
# size of our vectors
collection_name='imagesearch'
param = {'collection_name':collection_name, 'dimension':2048, 'metric_type':MetricType.L2}
milvus.create_collection(param)
```

```python
_, collection = milvus.get_collection_info(collection_name)
print(collection)
```

```python
# milvus takes 2d array as input
feature_vecs = np.concatenate(feature_vecs, axis=0 )
```

```python
# insert feature vectors into collection
status, ids = milvus.insert(collection_name=collection_name, records=feature_vecs)
```

```python
# create lookup dictionary by id for ease of plotting
# from David Shvimer
lookup = {}
for ID, img in zip(ids, images):
    lookup[ID] = img
```

```python
# flush features to disk
milvus.flush([collection_name])
# get collection row count i.e. # of entries
status, result = milvus.count_entities(collection_name)
# present collection statistics info
_, info = milvus.get_collection_stats(collection_name)
print(info)
```

```python
# prepare search paramaters and begin search
search_param = {
    "nprobe": 16
}
param = {
    'collection_name': collection_name,
    # feature vector of interest (we will use the same one)
    'query_records': np.reshape(feature_vecs[-4], (1, 2048,)),
    'top_k': 10,
    'params': search_param,
}
status, results = milvus.search(**param)
```

```python
# plot results
for neighbors in results:
    for n in neighbors:
        plt.figure()
        plt.subplot(1, 2, 1)
        plt.imshow(images[-4])
        plt.subplot(1, 2, 2)
        plt.imshow(lookup[n.id])
        print('Distance: {}'.format(n.distance))
```

While the distances for each pair are quite close, similar to the cosine similarities, the results from using Milvus with L2 distance offered an improvement over the prior task. There are no dogs in the top 10 anymore.
